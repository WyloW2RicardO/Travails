# Correlations

```python
from numpy import arange
from pandas import read_csv, to_datetime
from bokeh.plotting import figure, show
from bokeh.io import reset_output, output_notebook

reset_output()
output_notebook()
```

```python
nom_fichier = 'Q_33_previous-1950-2022_RR-T-Vent'
nom_var = 'TM'
nom_index = 'DATE'
nom_periode = 'Periode'
#nom_unit = 'D'
nom_format = '%Y%m%d'
nom_acf = 'ACF'
nombre_periode = 24 # 2 ans
val_mois = 12
```

```python
data_get = read_csv(nom_fichier + '_lieu.csv')
data_get
```

```python
data_format = data_get.copy()
data_format[nom_index] = to_datetime(
    data_format[nom_index]
    , format = nom_format
    # , unit = nom_unit
)
data_index = data_format.set_index(nom_index)
data_index
```

```python
df = data_index.copy()

graph = figure(
    x_axis_type="datetime"
)

graph.vbar(
    x = df.index
    , top = list(df[nom_var])
)

show(graph)
```

```python
#data_mois = data_index.groupby(data_index.index.month).mean()
data_mois = data_index.groupby(data_index.index.month).agg({'TM':['mean','std']})
data_mois.columns = [
    nom_var + '_mean'
    , nom_var + '_std'
]
data_mois
```

```python
df = data_mois.copy()

graph = figure()

graph.vbar(
    x = df.index
    , top = list(df[nom_var+"_mean"])
)

show(graph)
```

```python
#data_heur = data_index.loc[data_index.index.month == val_mois]
#data_heur = data_heur.groupby(data_heur.index.to_period('D')).mean()
#data_heur = data_heur.groupby(data_heur.index.day).mean()
data_heur = data_index.groupby(data_index.index.day).agg({'TM':['mean','std']})
data_heur.columns = [
    nom_var + '_mean'
    , nom_var + '_std'
]

data_heur
```

```python
df = data_heur.copy()

graph = figure(
    #x_axis_type="datetime"
)

graph.vbar(
    x = df.index
    , top = list(df[nom_var+"_mean"])
)

show(graph)
```

## Correlograme

```python
#data_date = data_index.groupby(data_index.index.to_period('M')).sum()
data_date = data_index.groupby(data_index.index.to_period('M')).agg({'TM':['mean','std']})
data_date.columns = [
    nom_var + '_mean'
    , nom_var + '_std'
]
data_date.head()
```

```python
range_period = arange(nombre_periode)
data_cor = list([data_date[nom_var + '_mean'].corr(data_date[nom_var + '_mean'].shift(index)) for index in range_period])
```

```python
graphique = figure()

graphique.vbar(
    x = range_period
    , top = data_cor
)

show(graphique)
```
