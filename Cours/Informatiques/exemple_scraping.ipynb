```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 1. Télécharger le contenu HTML de la page Wikipedia
url = "https://en.wikipedia.org/wiki/List_of_countries_by_life_expectancy"
response = requests.get(url)
html_content = response.content

# 2. Utiliser BeautifulSoup pour extraire les données du tableau
soup = BeautifulSoup(html_content, "html.parser")
table = soup.find("table", class_="wikitable")

# 3. Créer un DataFrame pandas à partir des données extraites
rows = table.find_all("tr")
data = []
for row in rows[1:]:
    cols = row.find_all(["th", "td"])
    cols = [col.text.strip() for col in cols]
    data.append(cols)

columns = ["Country", "Overall", "Male", "Female"]
df = pd.DataFrame(data, columns=columns)

# 4. Enregistrer le DataFrame dans un fichier CSV
df.to_csv("life_expectancy.csv", index=False)

print("DataFrame créé et enregistré avec succès.")
```
